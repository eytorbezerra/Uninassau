```python
#Jogo da Forca

import random

# Lista de palavras
palavras = ['maçã', 'tomate', 'uva', 'melancia', 'melão', 'laranja']

# Escolha aleatória da palavra
palavra = random.choice(palavras)

# Inicializa a lista de letras adivinhadas
letras_adivinhadas = ['_'] * len(palavra)

# Número de tentativas
tentativas_maximas = 5
tentativas = 0

# Inicio do joogo
while tentativas < tentativas_maximas:


    #Tema do Jogo
    print('---Frutas---')

    # Andamento do Jogo
    print(' '.join(letras_adivinhadas))

    # Escolha de uma Letra
    letra = input('Digite uma letra: ').lower()

    # Verifica se a letra está correta
    if letra in palavra:
        # Atualiza a lista de letras
        for i in range(len(palavra)):
            if palavra[i] == letra:
                letras_adivinhadas[i] = letra
    else:
        tentativas += 1
        print(f'Letra "{letra}" não está na palavra. Tentativas restantes: {tentativas_maximas - tentativas}')

    # Verifica se o Jogador Ganhou
    if '_' not in letras_adivinhadas:
        print('Parabéns! Você ganhou!')
        break

# Se Esgotar as Tentativas
if tentativas == tentativas_maximas:
    print(f'Você perdeu! A palavra era: {palavra}')
```

```python
#Simulador de Caixa Eletônico

#Saldo Ilustrativo
class CaixaEletronico:
    def __init__(self):
        self.saldo = 1400

#Funcionalidades do Caixa
    def consultar_saldo(self):
        return self.saldo

    def depositar(self, valor):
        self.saldo += valor
        return f'Depósito de R${valor} realizado. Saldo Atual: R${self.saldo}'

    def sacar(self, valor):
        if self.saldo >= valor:
            self.saldo -= valor
            return f'Saque de R${valor} realizado. Saldo Atual: R${self.saldo}'
        else:
            return 'Saldo insuficiente. Tente novamente mais tarde.'

# Menu
def exibir_menu():
    print("\nEscolha a opção:")
    print("1 - Consultar Saldo")
    print("2 - Depositar")
    print("3 - Sacar")
    print("4 - Sair")

# Criar uma instância do Caixa Eletrônico
caixa = CaixaEletronico()

# Loop das Opções
while True:
    exibir_menu()
    escolha = input("Digite o número da opção desejada: ")

    if escolha == '1':
        print(f"Saldo atual: R${caixa.consultar_saldo()}")
    elif escolha == '2':
        valor_deposito = float(input("Digite o valor a depositar: "))
        print(caixa.depositar(valor_deposito))
    elif escolha == '3':
        valor_saque = float(input("Digite o valor a sacar: "))
        print(caixa.sacar(valor_saque))
    elif escolha == '4':
        print("Obrigado por testar o caixa eletrônico. Até logo!")
        break
    else:
        print("Opção inválida. Por favor, escolha uma opção válida.")
```

```python
#Conversor de Moedas

# Definição dos Valores da Moedas
def obter_taxas_de_cambio():
    return {
        'USD': 1.0,
        'EUR': 0.85,
        'BRL': 5.05,

    }

def converter_moeda(valor, moeda_original, moeda_conversao, taxas):
    if moeda_original not in taxas or moeda_conversao not in taxas:
        return "Moeda não suportada."

    valor_em_usd = valor / taxas[moeda_original]
    valor_convertido = valor_em_usd * taxas[moeda_conversao]
    return valor_convertido

# Taxas de cambio
taxas = obter_taxas_de_cambio()

# Listar as moedas disponíveis
print("Moedas disponíveis:")
for moeda in taxas:
    print(moeda)

# Solicitações ao usuario
moeda_original = input("Digite a moeda de origem (ex: USD): ").upper()
moeda_conversao = input("Digite a moeda de destino (ex: EUR): ").upper()
valor = float(input("Digite o valor a ser convertido: "))

# Converção
resultado = converter_moeda(valor, moeda_original, moeda_conversao, taxas)
print(f'{valor} {moeda_original} equivale a {resultado} {moeda_conversao}')
```

```python
#Aplicativo de Previsão do Tempo

import requests

#Inicio do Funcionamento da API
def obter_previsao_tempo(cidade, chave_api):
    url = f'http://api.openweathermap.org/data/2.5/weather?q={cidade}&appid={chave_api}&units=metric'
    resposta = requests.get(url)

    if resposta.status_code == 200:
        dados = resposta.json()
        return dados
    else:
        return None

def exibir_previsao(dados):
    if dados is not None:
        nome_cidade = dados['name']
        descricao_clima = dados['weather'][0]['description']
        temperatura = dados['main']['temp']
        sensacao_termica = dados['main']['feels_like']

        print(f'-----Previsão do tempo para {nome_cidade}-----')
        print(f'Condição: {descricao_clima}')
        print(f'Temperatura: {temperatura}°C')
        print(f'Sensação térmica: {sensacao_termica}°C')
    else:
        print('Não foi possível obter a previsão do tempo.')
#Chave e Dados Necessario para a API
chave_api = '1d9fc560acfc275c48543324a370cf61'
cidade = 'Recife, BR'

dados_previsao = obter_previsao_tempo(cidade, chave_api)
exibir_previsao(dados_previsao)
```

```python
#Sistema de Reserva do Cinema

class Filme:
    def __init__(self, titulo, duracao):
        self.titulo = titulo
        self.duracao = duracao

class Sessao:
    def __init__(self, filme, horario, sala, vagas_disponiveis):
        self.filme = filme
        self.horario = horario
        self.sala = sala
        self.vagas_disponiveis = vagas_disponiveis

    def comprar_ingresso(self, quantidade):
        if self.vagas_disponiveis >= quantidade:
            self.vagas_disponiveis -= quantidade
            return f'Compra realizada com sucesso para {self.filme.titulo}. Total: R${quantidade*10}.'
        else:
            return 'Desculpe, não há vagas suficientes para a quantidade desejada.'

class Cinema:
    def __init__(self):
        self.sessoes = []

    def adicionar_sessao(self, sessao):
        self.sessoes.append(sessao)

    def listar_sessoes(self):
        for i, sessao in enumerate(self.sessoes, start=1):
            print(f'{i}. Filme: {sessao.filme.titulo}, Horário: {sessao.horario}, Sala: {sessao.sala}, Vagas Disponíveis: {sessao.vagas_disponiveis}')

# Filmes
filme1 = Filme("Elementos", "1h40m")
filme2 = Filme("Homem-Aranha: Através do Aranhaverso", "2h20m")

# Sessões
sessao1 = Sessao(filme1, "14:00", "Sala 1", 50)
sessao2 = Sessao(filme2, "16:30", "Sala 2", 30)

cinema = Cinema()
cinema.adicionar_sessao(sessao1)
cinema.adicionar_sessao(sessao2)

# Compra de ingressos
cinema.listar_sessoes()

escolha = int(input("Escolha o número da sessão desejada: "))
sessao_escolhida = cinema.sessoes[escolha - 1]

quantidade = int(input(f"Quantos ingressos para '{sessao_escolhida.filme.titulo}' você deseja comprar? "))
print(sessao_escolhida.comprar_ingresso(quantidade))

# Atualizar a lista de sessões
cinema.listar_sessoes()
```
